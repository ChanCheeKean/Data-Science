# Classification

## Logistic regression

```python
# define x and y
dataset = pd.read_csv('./data/Social_Network_Ads.csv')
x = dataset.iloc[:, [2,3]]
y = dataset.iloc[:, 4]

# splitting
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.25, random_state=0)
```

```python
# feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
# train model
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, Y_train)
```

```python
# predictions
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
# evaluation
cm = confusion_matrix(Y_test, y_pred)
pd.DataFrame(cm, columns = ['Predicted -', 'Predicted +'], index = ['Actual -', 'Actual +'])
```

```python
# to get probability of the dataset 
classifier.predict_proba(X_test)[:,1]
```

```python
# Visualising the Test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
```

## K Nearest Neighbours

### Data Import

```python
df = pd.read_csv("./data/Classified Data",index_col=0)
df.head()
```

### Scale and Transform

```python
# scale and transform
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('TARGET CLASS',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

### Splitting and Training

```python
# splitting
X_train, X_test, y_train, y_test = train_test_split(scaled_features, df['TARGET CLASS'], test_size=0.30)
```

```python
# Start with k =1
# prediction
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

### Evaluation

```python
cm = confusion_matrix(Y_test, y_pred)
print(pd.DataFrame(cm, columns = ['Predicted -', 'Predicted +'], index = ['Actual -', 'Actual +']))
print(classification_report(y_test,pred))
```

### Determination of K-Value

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i!= y_test))
```

```python
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

Bad Result, use GridSearch for better Parameters!

### GridSearch

Finding the right parameters (like what C or gamma values to use) 

GridSearchCV takes a dictionary that describes the parameters that should be tried and a model to train. The grid of parameters is defined as a dictionary, where the keys are the parameters and the values are the settings to be tested.

High c, low bias high variance
Large gamma, high bias, low variance

```python
# verbose define amount of messages
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=0)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
cm = confusion_matrix(y_test, grid_predictions)
print(pd.DataFrame(cm, columns = ['Predicted -', 'Predicted +'], index = ['Actual -', 'Actual +']))
print(classification_report(y_test,grid_predictions))
```

# Clustering

## K Means Clustering

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. 

Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. 

Distance calculated from centroid to centroid

Elbow method to choose possible K! 

### Data Import

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,random_state=101)
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

### Creating the Clusters

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

### KMeans Example - 2 

```python
dataset = pd.read_csv('./data/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]]
```

```python
# Using the elbow method to find the optimal number of clusters
# kmeans.inertia to find wcss/total distance from point to centroid 
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# plotting
plt.plot(range(1, 11), wcss)
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter=300, n_init=10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=y_kmeans,cmap='rainbow');
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
```

## Hierachy Clustering

Different from KMeanClustering

Using dendogram to find the optimal number of cluster

Largest distance at dendogram (from one line to next), draw a line to decide no. of cluster

```python
# Importing the dataset
dataset = pd.read_csv('./data/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]]
```

```python
# cluster = 5 for this case
# ward linkage to minimize variance in each cluster
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
```

```python
hc = AgglomerativeClustering (n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=y_hc, cmap='rainbow');
```
